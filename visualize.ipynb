```python
def ctshow(image_arr):
    plt.close()
    plt.imshow(image_arr, cmap='gray', dpi=300)#, vmin=0, vmax=1)
    plt.show()
    
def ctsave(image_arr, out_path):
    plt.close()
    plt.imshow(image_arr, cmap='gray')#, vmin=0, vmax=1)
    plt.savefig(out_path, dpi=300)    

def ctload(image_path):
    with open(image_path, 'rb') as f:
        tif = pil_image.open(io.BytesIO(f.read()))
    return np.array(tif)
```

```python
# path=r"covid_data/patient1/SR_3/IM00001.tif"
scan_dir = r"covid_data/patient1/SR_2/"

import os

for f_name in sorted(os.listdir(scan_dir)):
    f_path = os.path.join(scan_dir, f_name)

    arr = ctload(f_path)

    ctshow(arr)
```
